# 目標1： 行列とは ― 行列とスカラーの四則演算を行う

## 1. 配列を作成し、配列とスカラーの四則演算を行うコード

```python
import numpy as np

A = np.array([[379.7, 350.7, 1154.4],
             [284.6, 269.2,  771.3]])

print(A)
print(A[0, 0])
print(A[1, 2])
```

# 目標2： 配列同士の四則演算を行う

## 2. 配列同士の四則演算を行うコード

```python
import numpy as np

A = np.array([[379.7, 350.7, 1154.4],
             [284.6, 269.2,  771.3]])
B = np.array([[10,   0, 0],
              [20, 100, 0]])

print(A + B)
```

```python
import numpy as np

image = np.array([[255, 24, 10],
                  [31, 240, 2],
                  [10, 8, 243]])
mask = np.eye(3)  # 3行3列の単位行列を作る

print((image * mask).mean())
print((image * np.fliplr(mask)).mean())  # maskを左右反転
```

# 目標4： 一部の行や列を取り出す、行や列の合計／平均を求める

## 4. 一部の行や列を取り出す（スライス機能）

```python
import numpy as np

A = np.array([[95, 94, 81, 90, 91],
              [85, 90, 87, 71, 83],
              [74, 98, 93, 74, 87],
              [88, 87, 90, 94, 91]])

print(A[1, :])
# 出力例：
# [85 90 87 71 83]  # 1行目のデータ

print(A[:, 1:3])  # 1列目から3列目の手前まで（1列目と2列目)
# [[94 81]
#  [90 87]
#  [98 93]
#  [87 90]]  # 1列目と2列目のデータ（理数系科目）

print(A[:, (0, 3, 4)])
# [[95 90 91]
#  [85 71 83]
#  [74 74 87]
#  [88 94 91]]  # 0列目、3列目、4列目のデータ（文科系科目）
```

## 5. 行や列の合計・平均を求める

```python
import numpy as np

A = np.array([[95, 94, 81, 90, 91],
              [85, 90, 87, 71, 83],
              [74, 98, 93, 74, 87],
              [88, 87, 90, 94, 91]])

print(np.sum(A))  # 全体の合計
print(np.mean(A, axis=1))  # 各行の平均
print(np.std(A, axis=0))  # 各列の標準偏差
# 出力例：
# 1743  # 全体の合計
# [90.2 83.2 85.2 90. ]  # 個人ごとの平均（Hさんの平均＝90.2など）
# [7.56637298 4.14578099 4.43705984 9.90896059 3.31662479]  # 科目ごとの標準偏差（国語の標準偏差＝7.56……など）
```

## （1）共通鍵方式によって画像を暗号化／復号する

image.png

```python
from imageio import imread
import matplotlib.pyplot as plt

url = 'https://github.com/Gessys/math/blob/main/data/bird.png?raw=true'
image = imread(url)  # imageに読み込む
plt.imshow(image)  # 表示してみる
plt.show()
```

```python
from imageio import imread
import numpy as np
import matplotlib.pyplot as plt

# 加工前（オリジナル画像）
print("加工前（オリジナル画像）")
url = 'https://github.com/Gessys/math/blob/main/data/bird.png?raw=true'
image = imread(url)  # imageに読み込む
plt.imshow(image)  # 表示してみる
plt.show()

# 送信側
print("送信側（暗号化）")
np.random.seed(255) # 共通鍵の指定（乱数の種である255が共通鍵）
# 擬似乱数の作成（画像データと同じ個数）
r = np.random.randint(0, 256, size=image.size, dtype=np.uint8)
mask = r.reshape(image.shape) # 作成された乱数をイメージと同じ形の配列にする
encrypted = image ^ mask # 排他的論理和を求め暗号化する
plt.imshow(encrypted)  # 表示してみる
plt.show()

# 受信側
print("受信側（複合化）")
np.random.seed(255) # 共通鍵の指定（乱数の種である255が共通鍵）
# 擬似乱数の作成（画像データと同じ個数）
r = np.random.randint(0, 256, size=image.size, dtype=np.uint8)
mask = r.reshape(image.shape) # 作成された乱数をイメージと同じ形の配列にする
encrypted = encrypted ^ mask # 排他的論理和を求め暗号化する
plt.imshow(encrypted)  # 表示してみる
plt.show()
```

```python
from imageio import imread
import numpy as np
import matplotlib.pyplot as plt

# 加工前（オリジナル画像）
print("加工前（オリジナル画像）")
url = 'https://github.com/Gessys/math/blob/main/data/bird.png?raw=true'
image = imread(url)
plt.imshow(image)
plt.show()

def decomp(image, mask):
  return image * mask

# 加工後（Gとαを取り出す）
print("加工後（Gとαを取り出す）")
mask = [0, 1, 0, 1] # Gとαを取り出す
gch = decomp(image, mask)
plt.imshow(gch)
plt.show()
```

```python
from imageio import imread
import numpy as np
import matplotlib.pyplot as plt

# 加工前（オリジナル画像）
print("加工前（オリジナル画像）")
url = 'https://github.com/Gessys/math/blob/main/data/bird.png?raw=true'
image = imread(url)
plt.imshow(image)
plt.show()

def magcnv(image, mag):
  return np.asarray((image * mag).clip(max=255), dtype=np.uint8)

# 加工後（mag倍明るくする）
print("加工後（mag倍明るくする）")
gch = magcnv(image, 2.0)
plt.imshow(gch)
plt.show()
```
